```python

sns.set_style('whitegrid')
lista1 = list()
mis_etiquetas =list()

for gen in peliculas.Genero.cat.categories:
    lista1.append(peliculas[peliculas.Genero == gen].PresupuestoMillones)
    mis_etiquetas.append(gen)

fig, ax=plt.subplots(1,1)
fig.set_size_inches(11.7, 8.27) #Tamaño de una hoja A4

# Colores para cada categoría
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'darkblue']  # Ejemplo de lista de colores

h = plt.hist(lista1, bins=30, stacked=True, rwidth=1, edgecolor='none', label=mis_etiquetas, color=colors)
plt.title('Distribución de Presupuestos para las Películas', \
          fontsize=35, color='DarkBlue', fontname='Arial')
plt.ylabel('Número de Películas', fontsize=25, color='Red')
plt.xlabel('Presupuesto', fontsize=25, color='Green')
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(frameon=True, fancybox=True, framealpha=1, shadow=True, prop={'size':20})
plt.show()
```
